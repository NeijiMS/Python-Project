## Data Wrangling Process

```python
# Importing Pandas library and instructing Pandas to read the dataset

import pandas as pd

mydata = pd.read_csv('teleCust1000t.csv')

print ('done')
```

```python
# Renaming the columns' headers

headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
print("headers\n", headers)
```

```python
# Rechecking the 'renaming' progress

mydata.columns = headers
mydata.head(5)
```

```python
# Importing numpy library
# Finding and replacing missing values using numpy

import numpy as np

mydata.replace("?", np.nan, inplace = True)
mydata.head(30)
```

```python
# Previous function's continuation (replacing NaN to True)

missing_data = mydata.isnull()
missing_data.head(30)
```

```python
# Using loop to count the missing values (True) in each column

for column in missing_data.columns.values.tolist(): 
    print(column)
    print (missing_data[column].value_counts())
    print("")
```

Based on the summary above, each column has 204 rows of data, seven columns containing missing data:

<ol>
    <li>"normalized-losses": 41 missing data</li>
    <li>"num-of-doors": 2 missing data</li>
    <li>"bore": 4 missing data</li>
    <li>"stroke" : 4 missing data</li>
    <li>"horsepower": 2 missing data</li>
    <li>"peak-rpm": 2 missing data</li>
    <li>"price": 4 missing data</li>
</ol>

Whole columns should be dropped only if most entries in the column are empty. In this dataset, none of the columns are empty enough to drop entirely.
I have some freedom in choosing which method to replace data; however, some methods may seem more reasonable than others. I will apply each method to many different columns:

<b>Replace by mean:</b>

<ul>
    <li>"normalized-losses": 41 missing data, replace them with mean</li>
    <li>"stroke": 4 missing data, replace them with mean</li>
    <li>"bore": 4 missing data, replace them with mean</li>
    <li>"horsepower": 2 missing data, replace them with mean</li>
    <li>"peak-rpm": 2 missing data, replace them with mean</li>
</ul>

<b>Replace by frequency:</b>

<ul>
    <li>"num-of-doors": 2 missing data, replace them with "four". 
        <ul>
            <li>Reason: 84% sedans is four doors. Since four doors is most frequent, it is most likely to occur</li>
        </ul>
    </li>
</ul>

<b>Drop the whole row:</b>

<ul>
    <li>"price": 4 missing data, simply delete the whole row
        <ul>
            <li>Reason: price is what I want to predict. Any data entry without price data cannot be used for prediction; therefore any row now without price data is not useful</li>
        </ul>
    </li>
</ul>

```python
# Calculating the mean values for the column to be replaced

mean_norm_loss = mydata["normalized-losses"].astype("float").mean(axis=0)
print("Mean value of normalized-losses:", mean_norm_loss)

mean_stroke = mydata["stroke"].astype("float").mean(axis=0)
print("Mean value of stroke:", mean_stroke)

mean_bore = mydata["bore"].astype("float").mean(axis=0)
print("Mean value of bore:", mean_bore)

mean_horsepower = mydata["horsepower"].astype("float").mean(axis=0)
print("Mean value of horsepower:", mean_horsepower)

mean_peak_rpm = mydata["peak-rpm"].astype("float").mean(axis=0)
print("Mean value of peak_rpm:", mean_peak_rpm)
```

```python
# Replacing 'True' with mean value in for the related column

mydata["normalized-losses"].replace(np.nan, mean_norm_loss, inplace=True)
mydata["stroke"].replace(np.nan, mean_stroke, inplace=True)
mydata["bore"].replace(np.nan, mean_bore, inplace=True)
mydata["horsepower"].replace(np.nan, mean_horsepower, inplace=True)
mydata["peak-rpm"].replace(np.nan, mean_peak_rpm, inplace=True)
print ('done')
```

```python
# Checking the replacement progress

mydata.head(30)
```

```python
# Finding the frequency for the column to be replaced

mydata['num-of-doors'].value_counts()
```

```python
# Finding the max frequency for the column to be replaced

mydata['num-of-doors'].value_counts().idxmax()
```

```python
# Finding the min frequency for the column to be replaced

mydata['num-of-doors'].value_counts().idxmin()
```

```python
# Replacing 'True' with max frequency for the related column

mydata['num-of-doors'].replace(np.nan, 'four', inplace=True)
print ('done')
```

```python
# Droping the whole row that has missing value

mydata.dropna(subset=['price'], axis=0, inplace=True)
print ('done')
```

```python
# Checking the data size again after some cleaning

mydata.shape
```

```python
# Checking the existance of missing value (True)

after_missing_data = mydata.isnull()
after_missing_data.head(5)
```

```python
# Using loop again to count the missing values (True) in each column

for column in after_missing_data.columns.values.tolist(): #this for loop to count the values in each column
    print(column)
    print (after_missing_data[column].value_counts())
    print("")
```

```python
# Now after the data is cleaned (no missing value), we can change the types of the data

mydata.dtypes
```

```python
# Changing the data types for related columns

mydata[['normalized-losses']] = mydata[['normalized-losses']].astype('int')
mydata[['bore']] = mydata[['bore']].astype('float')
mydata[['stroke']] = mydata[['stroke']].astype('float')
mydata[['price', 'peak-rpm']] = mydata[['price', 'peak-rpm']].astype('float')
mydata[['horsepower']] = mydata[['horsepower']].astype('int')
print ('done')
```

```python
# Checking the replacement progress

mydata.dtypes
```

```python
# Changing the name of the existed columns

mydata['city-L/100km'] = 235/mydata['city-mpg']
mydata['highway-L/100km'] = 235/mydata['highway-mpg']
print ('done')
```

```python
# Checking the replacement progress

mydata.head(5)
```

```python
# Implementing Data Normalization (using max value) for the related columns

mydata['width'] = mydata['width']/mydata['width'].max()
mydata['height'] = mydata['height']/mydata['height'].max()
mydata['length'] = mydata['length']/mydata['length'].max()
print ('done')
```

```python
# Next, we're going to binning (classify into categories) 'Horsepower' column 
# First, call out the column for understanding

mydata['horsepower'].head(30)
```

```python
# Importing Matplotlib library to visualize the dataset
# The visualization isn't in clear manner 

%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
plt.pyplot.hist(mydata["horsepower"])

# set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

```python
# Using linspace function in numpy library to bin the data

bins = np.linspace(min(mydata["horsepower"]), max(mydata["horsepower"]), 4)
bins
```

```python
# Innitiating a list for the binning

categories_names = ['Low', 'Medium', 'High']
print ('done')
```

```python
# Converting the cateorical values to numerical values (binning)

mydata['horsepower-binned'] = pd.cut(mydata['horsepower'], bins, labels=categories_names, include_lowest=True )
mydata[['horsepower','horsepower-binned']].head(20)
```

```python
# Finding the frequency for the column

mydata['horsepower-binned'].value_counts()
```

```python
# Initiating matplotlib again after the binning process
# The visualization is clearer

%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
pyplot.bar(categories_names, mydata["horsepower-binned"].value_counts())

# set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

```python
# Innitiating Dummies to convert categorical column (with only 2 categories) to numerical columns
# The first numerical column to be converted is; 'Fuel Type'

dummies1 = pd.get_dummies(mydata['fuel-type'])
dummies1.tail(30)
```

```python
# Renaming the dummies' columns

dummies1.rename(columns={'fuel-type-diesel':'gas', 'fuel-type-diesel':'diesel'}, inplace=True)
print ('done')
```

```python
# Merging data frame 'mydata' and 'dummies_1' 

mydata = pd.concat([mydata, dummies1], axis=1)
mydata.head(5)
```

```python
# Droping the original column 'fuel-type' from 'mydata'

mydata.drop('fuel-type', axis = 1, inplace=True)
mydata.head(5)
```

```python
# Innitiating Dummies to convert categorical column (with only 2 categories) to numerical columns
# The second numerical column to be converted is; 'Aspiration'

dummies2 = pd.get_dummies(mydata['aspiration'])
dummies2.tail(30)
```

```python
# Renaming the dummies' columns

dummies2.rename(columns={'std':'aspiration-std', 'turbo':'aspiration-turbo'}, inplace=True)
dummies2.tail(30)
```

```python
# Merging data frame 'mydata' and 'dummies 2' 

mydata = pd.concat([mydata, dummies2], axis=1)
mydata.head(5)
```

```python
# Droping the original column 'aspiration' from 'mydata'

mydata.drop('aspiration', axis = 1, inplace=True)
mydata.head(5)
```

```python
# Saving the cleaned dataset to a new file

mydata.to_csv('cleanedcarpricing.csv')
```
